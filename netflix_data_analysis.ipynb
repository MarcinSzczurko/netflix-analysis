# Netflix data analysis

## Libraries

```python
import pandas as pd
```

## Load data

```python
df = pd.read_csv('netflix_titles.csv')
df.info()
df
```

## Data transformation

```python
# Sort table by release_year 
df.sort_values('release_year', ascending=False, inplace=True)
df
```

```python
# Fill in missing data
df['director'].fillna('Unknow', inplace=True)
df['country'].fillna('Unknow', inplace=True)
```

## German cinematography 

### Fully German productions

```python
# Select fully German productions released after 2010 
german_productions_full_df = df[(df['country'] == 'Germany') & (df['release_year'] > 2010)]
german_productions_full_df
```

### All German productions

```python
# Select all German productions released after 2010 
german_productions_all_df = df[(df['country'].str.contains('Germany')) & (df['release_year'] > 2010)]
german_productions_all_df
```

### Productions with German participation

```python
german_productions_participation_df = german_productions_all_df.merge(german_productions_full_df[['show_id', 'title']], on='show_id', how='left')
german_productions_participation_df = german_productions_participation_df[german_productions_participation_df['title_y'].isnull()]\
                                                                                                        .drop('title_y', axis='columns')\
                                                                                                        .rename({'title_x' : 'title'}, axis='columns')
german_productions_participation_df
```

## Swedish cinematography

```python
# Show all productions where Sweden was only country or main country of production (listed as first)
mainly_sweden_films = df.copy()

mainly_sweden_films['main_country'] = mainly_sweden_films['country'].apply(lambda x: x.split(',')[0] if ',' in x else x)

mainly_sweden_films = mainly_sweden_films[mainly_sweden_films['main_country'] == 'Sweden']
mainly_sweden_films
```

## Cinematic productions by country and year (pivot table)

```python
# Support table (preview)
grouped_df = df.groupby(['type', 'release_year'])\
                ['show_id'].count().reset_index()
grouped_df
```

```python
# Clean values of date_added columns and change type to datetime
df['date_added'] = df['date_added'].str.strip()
df['date_added'] = pd.to_datetime(df['date_added'])
```

```python
# Group DataFrame by type and release_year 
# Count show_id and show min date_added 
grouped_df = df.groupby(['type', 'release_year'])\
               .agg({'show_id': 'count', 
                     'date_added': 'min'})
print(grouped_df.to_string())
```

```python
# Create pivot DataFrame: show_id count by release_year and main_country
df['main_country'] = df['country'].apply(lambda x: x.split(',')[0] if ',' in x else x)
country_year_pivot_df = pd.pivot_table(df, columns='release_year', index='main_country', values='show_id', aggfunc='count')

# Fill missing data with 0
# Reverse columns order and change type to int
country_year_pivot_df = country_year_pivot_df.fillna(0)
country_year_pivot_df = country_year_pivot_df[reversed(country_year_pivot_df.columns)].astype(int)

# Add column sum (sum of rows) and sort values by new column
country_year_pivot_df['sum'] = country_year_pivot_df.sum(axis='columns')
country_year_pivot_df = country_year_pivot_df.sort_values('sum', ascending=False)#.drop('sum', axis='columns')

country_year_pivot_df
```

## Movies average duration by country and year (pivot table)

```python
df['duration']
```

```python
# Select needed columns
# Filter only movies
# Extract the duration (minutes)
task_df = df[['type', 'release_year', 'main_country', 'duration']]
task_df = task_df[task_df['type'] == 'Movie'].reset_index(drop=True)
task_df['duration'] = task_df['duration'].apply(lambda x: int(x.split(' ')[0]))

# Create pivot DataFrame: average movie duration by release_year and main_country
# Fill in missing data with 0
pivot_df = pd.pivot_table(task_df, columns='release_year', index='main_country', values='duration', aggfunc='mean')
pivot_df = pivot_df.fillna(0)

# Reverse columns order and change type to int
pivot_df = pivot_df[reversed(pivot_df.columns)].astype(int)

pivot_df
```
